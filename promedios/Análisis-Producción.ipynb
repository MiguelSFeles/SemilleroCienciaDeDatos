```python
Toneladas = pd.DataFrame()
Toneladas['Departamento'] = None
Toneladas['Produccion'] = None

for departamento in agro.Departamento :
    if departamento in list(Toneladas.Departamento): continue
    suma = round(sum(agro[agro.Departamento == departamento]['Producción_(t)']))
    agregado = {'Departamento': departamento, 'Produccion': suma}
    Toneladas = Toneladas.append(agregado, ignore_index=True)

Toneladas.sort_values('Produccion', ascending=False)
```

```python
CosechaArea = pd.DataFrame()
CosechaArea['Departamento'] = None
CosechaArea['Cosecha'] = None

for departamento in agro.Departamento :
    if departamento in list(CosechaArea.Departamento): continue
    suma_cos = round(sum(agro[agro.Departamento == departamento]['Área_cosechada_(ha)']))
    agregado2 = {'Departamento': departamento, 'Cosecha': suma_cos}
    CosechaArea = CosechaArea.append(agregado2, ignore_index=True)

CosechaArea.sort_values('Cosecha', ascending=False)
```

```python
Alimentos = pd.DataFrame()
Alimentos['Departamento'] = None
Alimentos['Alimentos'] = None

for departamento in agro.Departamento :
    if departamento in list(Alimentos.Departamento): continue
    lenAlim = len(agro[agro.Departamento == departamento]['Sistema_Productivo'].unique())
    agregado3 = {'Departamento': departamento, 'Alimentos': lenAlim}
    Alimentos = Alimentos.append(agregado3, ignore_index=True)

Alimentos.sort_values('Alimentos', ascending=False)
```
