```python
agro = evalAgro = pd.read_csv('EvaluaciónAgropecuariaClean.csv')
agro = agro.drop('Unnamed: 0', axis=1)#Borré el index
agro.info()
```

```python
Antioquia = agro[agro.Departamento=="Antioquia"]
t_casanare = sns.lineplot(data=Antioquia.groupby(["Año",'Producción_(t)','Estado_Físico_de_Producción']).size(), x="Año", y="Producción_(t)", hue="Estado_Físico_de_Producción",palette = "husl")
plt.title("Antioquia")
plt.show()
```
